```python
# Import libraries
import pandas as pd
```

```python
#Url data
url = './data/raw/PRECIOS LUZ.xlsx'
```

```python
# Extracting dataframe and preprocessing
df_fixed_price = pd.read_excel(url, sheet_name='FIJO')
df_fixed_price = df_fixed_price.iloc[1:,2:]
df_fixed_price.dropna(inplace=True)
df_fixed_price.columns = ['SISTEMA','CIA','PRODUCTO','PRODUCTO_CIA','TARIFA', 'FEE',
                          'P1','P2','P3','P4','P5','P6',
                          'P1.','P2.','P3.','P4.','P5.','P6.'
                            ]
```

```python
# Nice
df_fixed_price
```

```python
# Saving to csv
df_fixed_price.to_csv('./data/processed/fixed_price.csv', index=False)
```

```python
# Extracting dataframe df_indexed_price and preprocessing
df_indexed_price = pd.read_excel(url, sheet_name='INDEXADO')
```

```python
# Proper data and nice columns
df_indexed_price = df_indexed_price.iloc[3:,1:12]
df_indexed_price.columns=['SISTEMA','TARIFA','CIA','MES','FEE','P1','P2','P3','P4','P5','P6']
df_indexed_price.reset_index(drop=True, inplace=True)
```

```python
df_indexed_price
```

```python
# Saving to csv
df_indexed_price.dropna(inplace=True)
df_indexed_price.reset_index(drop=True, inplace=True)
df_indexed_price.to_csv('./data/processed/indexed_price.csv', index=False)
```

```python
# Extracting dataframe df_fixed_price_power and preprocessing

df_indexed_price_power = pd.read_excel(url, sheet_name='INDEXADO')
```

```python
# Making everything nice
df_indexed_price_power = df_indexed_price_power.iloc[3:, 13:]
df_indexed_price_power.columns = ['SISTEMA','CIA','PRODUCTO','PRODUCTO_CIA','TARIFA','P1','P2','P3','P4','P5','P6']
df_indexed_price_power.dropna(inplace=True)
df_indexed_price_power.reset_index(drop=True, inplace=True)
```

```python
df_indexed_price_power
```

```python
# Saving to csv
df_indexed_price_power.to_csv('./data/processed/indexed_price_power.csv', index=False)
```
